## AutoViz 라이브러리 활용하는 부분

```python
dataset = pd.read_csv("data/split/train/train_T100306_T31.csv")
dataset = dataset.drop(columns=['PRODUCT_ID', 'TIMESTAMP', 'Y_Class', 'LINE', 'PRODUCT_CODE'])
```

```python
# dataset = dataset.sample(200)
```

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
```

```python
%matplotlib inline

# 시각화 결과 저장 경로
save_plot_dir = "./data/split/train/eda_results"

# 자동 시각화 실행
dft = AV.AutoViz(
    filename="",
    sep=",",
    depVar="Y_Quality",
    dfte=dataset,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="html",
    max_rows_analyzed=200,
    max_cols_analyzed=700,
    save_plot_dir=save_plot_dir
    )
```
